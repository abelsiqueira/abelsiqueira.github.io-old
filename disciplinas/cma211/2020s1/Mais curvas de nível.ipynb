```julia
using Plots
gr(size=(400,300))
plot(rand(3),rand(3))
surface(rand(10),rand(10),(x,y)->x^2+y^2)
contour(rand(10),rand(10),(x,y)->x^2+y^2)
nothing
```

```julia
f(x,y) = x^2 + y^2  1 ? sqrt(1 - x^2 - y^2) : NaN
xg = range(-1.2, 1.2, length=400)
contour(xg, xg, f, levels=20, leg=false)
```

```julia
f(x,y) = x  0 ? 2x + 3y + 4 : 3x + 7 + y
xg = range(-2, 2, length=100)
contour(xg, xg, f, levels=51, leg=false)
```

```julia
f(x,y) = x / (y - 1)
xg = range(-2, 2, length=100)
contour(xg, xg, f, levels=[-2,-1,0,1,2], leg=false)
```

```julia
f(x,y) = (x^2 + y) / (x + 2)
xg = range(-4, 4, length=100)
yg = range(-8, 2, length=100)
contour(xg, yg, f, levels=-9:3, leg=false)
```
