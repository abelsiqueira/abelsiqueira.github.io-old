```julia
B = [4.0 1.0; 1.0 2.0]
g = -B*[1.2; 0.8] # Roubei

m(d) = 0.5*dot(d, B*d) + dot(d, g)
```

```julia
contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
(x,y)->m([x;y]))
```

```julia
d = -B\g
```

```julia
norm(d)
```

```julia
t = linspace(0, 2pi, 100)
@manipulate for  = linspace(0.1, 2, 100),
     = linspace(0, 10, 100)
    contour(linspace(-0.2, 2, 100), linspace(-0.2, 2, 100),
        (x,y)->m([x;y]))
    plot!(*cos(t), *sin(t), c=:blue, l=:dash)
    d = -(B + *eye(2))\g
    
    plot!([0, d[1]], [0, d[2]], c=:red)
    
    xlims!(-0.2, 2)
    ylims!(-0.2, 2)
end
```
